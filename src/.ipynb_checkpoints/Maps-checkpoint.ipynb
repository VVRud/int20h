```python
def prep_data_geocode(params, lat, lng):
    lst = []
    for i in range(len(lat)):
        par = params.copy()
        par['latlng'] = '{},{}'.format(lat[i], lng[i])
        lst.append(par)
    
    return lst

def prep_data_directions(params, o_lat, o_lng, d_lat, d_lng):
    lst = []
    for i in range(len(o_lat)):
        par = params.copy()
        par['origin'] = '{},{}'.format(o_lat[i], o_lng[i])
        par['destination'] = '{},{}'.format(d_lat[i], d_lng[i])
        
        lst.append(par)
    return lst
```

```python
def get_data(url, params):
    async def send_get(sess, id_, params_):
        async with sess.get(url, params=params_) as resp:
            return list([id_, await resp.json()])

    async def get_resp():
        resp_list = []

        async with ClientSession() as session:
            it = await asyncio.gather(*[
                        send_get(session, id_, params[id_])
                        for id_ in range(len(params))
            ])
            for responce in it:
                resp_list.append(responce)
        
        return resp_list
    res = sorted(asyncio.get_event_loop().run_until_complete(get_resp()), key=lambda x: x[0])
    return [result[1] for result in res]
```

```python
get_data(
    'https://maps.googleapis.com/maps/api/geocode/json',
    prep_data_geocode({
        'key': 'AIzaSyDLQGmmrDRpvZFUgGfkviBZ_iiR8TtWzSk'
    }, lat, lng))
```
